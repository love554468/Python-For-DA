```python
# 1.	Viết hàm để tính độ dài của 1 vector
import math
A, B = (1,10), (-30,29)

def d(A, B):
    return round(math.sqrt((A[0] - B[0])**2+(A[1] - B[1])**2),2)

d(A,B)
```

```python
# 2.	Viết hàm số để tính góc giữa hai vector
# asp dungt theo cong thuc tren
def goc_2vec(A, B):
    return (A[0]*B[0] + A[1]*B[1]) / (((A[0]**2+A[1]**2)**0.5) * ((B[0]**2+B[1]**2)**0.5))

goc_2vec((-1,1), (1,3))
```

```python
# 3.	Viết hàm để tính chu vi và diện tích của bất kỳ hình tròn nào khi biết bán kính
def tamgiac(r):
    chuvi = round(3.14*2*r,2) # 2pi r
    dientich = round(3.14*(r**2),2) # pi r^2
    print(f"C: {chuvi} \nS: {dientich}")

    
tamgiac(5)
```

```python
# 5.	Viết hàm tính trung bình của một list bất kỳ

def avg_(*arg):
    k = 0
    avg = 0
    
    for i in arg:
        k+=1
        avg+=i
        
    return avg/k

avg_(1,2,3,4)
```

```python
# 7.	Viết hàm tìm min và max của một list bất kỳ (không dùng min(), max())

def com_min(a, b):
    return a<b

def f_min(*args):
    re = float('inf')
    for i in args:
        if com_min(i, re):
            re = i
    return re

f_min(1,2,-1,10,10)

#tuong tu voi ham max
```

```python
def com_max(a, b):
    return a>b

def f_max(*args):
    re = float('-inf')
    for i in args:
        if com_max(i, re):
            re = i
    return re

f_max(1,2,-1,10,10)
```
