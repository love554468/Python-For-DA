## video 9.10: function: reduce

trong kĩ thuật người ta hay nhắn đến map và reduce

### (#) Loop and function approach

```python
# cách tiếp cận thủ công tìm max trong list các phần tử
l = [1,3,120,5,6,3,10]

max_ = l[0]

for i in l[1:]:
    if i> max_:
        max_ =i

max_
```

```python
# cách tiếp cận sài loop và function

print(l)

def max_val(x, y):
    return x if x> y else y


def max_seq(seq): # trả về max của 1 list
    re = seq[0]
    for i in seq[1:]:
        re = max_val(re, i)
    return re
    
max_seq(l)
```

```python
# cách tiếp cận dùng lambda và function apply - nâng cao
# sài cho nhiều function cùng lúc

max_val = lambda x, y: x if x>y else y
min_val = lambda x, y: x if x<y else y
sum_val = lambda x, y: x+y
mul_val = lambda x, y: x*y

def apply(fn, seq): # vẫn là trả về max hoặc min hoặc sum, mul nhưng mang ý
    # nghĩa của hàm reduce hơn chỉ là tính tổng hay max mul
    re = seq[0]
    for i in seq[1:]:
        re = fn(re, i)
    return re
    
apply(max_val, l)
```

```python
apply(min_val, l)
```

```python
apply(sum_val, l)
```

```python
apply(mul_val, l)
```

Ghi chú: về hàm apply mô tả tính chất của hàm reduce, thực chất hàm reduce là phép giảm chiều dữ liệu trả về 1 single value, áp dụng trên iterable sử dụng từng kết quả liền kề với nhau một để tính toán

```python
# global keyword

sum_ = 0
def f(*args):
    global sum_
    for i in args:
        sum_+=i
    return sum_

f(1,2,3,43)
```

## Enclosing scope

Chỉ tồn tại trong nested func

```python
def outer():
    x = 1
    
    def inner():
        y = 2
        print(x)
        print(y)
        
    inner()
    print(x)
    print(y)
    
outer()
```

we cannot access y from outer because y is not a scope of outer function

we can access x from inner because the scope of x is larger

### (#) Free variable

```python
# First ordered function

def f():
    x = 'local to f'
    print(x)
```

```python
# Nested function Without free variable

def outer():
    x = 'Local to f' # vẫn có enclosing scope nhưng 
    # không sài xuống dưới nhưng ngầm hiểu vẫn có
    
    def inner():
        x = 'Local to inner' # Not a free variable
        print(x)
    
    print(x)
    inner()
outer()
```

```python
# Nested function with free variable
def outer():
    x = 'local to outer'
    
    def inner():
        print(x, ": free variable in inner scope")
        # x refers to the string object in outer's scope
        # Nonlocal variable x is called a free variable
    
    inner()
outer()
```

Notes:
Khi hiểu về clourses

+ the function: inner
+ the free variable x

Variable x in outer's scope + function inner ảe bound together: clourses

clourses là gắn kết giữa một biến x ở outer và inner thông qua free variable x

clourses hợp của enclosing scope, nested function, và free variable

### Multiple intances of closures

Each time a function is generated, a new scope is created --> a new closure is created as well

```python
sum_ = 0
def f_sum(*args):
    global sum_
    for i in args:
        sum_ += i
    return sum_

f_sum(1,2,3,4)
```

```python
def outer(*args):
    sum_ = 0
    
    def inner():
        nonlocal sum_
        for i in args: # free variable (tuple object) - khong the thay doi gia tri
            sum_ += i # free variable - cos the thay doi gia tri
        
        return sum_
    return inner

outer(1,2,3)()
```

```python
f1 = outer(1,2,3)
f1
```

```python
f1.__closure__
```

```python
f1.__code__.co_freevars
```

```python
f2 = outer(1,2,3)
f2.__closure__
```

### Shared extended scope - chia se scope mo rong

```python
count = 0

def outer():
    count = 0
    
    def inner1():
        nonlocal count
        count +=1
        return count
    
    def inner2():
        nonlocal count
        count +=1
        return count
    
    return inner1, inner2
```

```python
f1, f2 = outer()

f1()
```

```python
f2()
```

### tại sao?
